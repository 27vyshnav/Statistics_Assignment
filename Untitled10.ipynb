**1.An Institution wishes to find out their student’s ability in mathematics, reading and writing skills. The Institution wants to do an exploratory study to check the following information.**

```python
#a.Find out how many males and females participated in the test.

gender_counts = data['gender'].value_counts()

print("Number of males:", gender_counts['male'])
print("Number of females:", gender_counts['female'])
```

```python
#b.What do you think about the students' parental level of education?

import matplotlib.pyplot as plt
parental_level_of_education = data['parental level of education'].value_counts()
parental_level_of_education
```

```python
#c.Who scores the most on average for math, reading and writing based on
#!.Gender
#!!.Test preparation course

gender_avg_scores = data.groupby('gender').mean()
prep_course_avg_scores = data.groupby('test preparation course').mean()
print("Average scores based on gender:\n", gender_avg_scores)
print("\nAverage scores based on test preparation course:\n", prep_course_avg_scores)
```

```python
#d.What do you think about the scoring variation for math, reading and writing based on
#!.Gender
#!!.Test preparation course

gender_std_scores = data.groupby('gender').std()

prep_course_std_scores = data.groupby('test preparation course').std()

print("Standard deviation of scores based on gender:\n", gender_std_scores)
print("\nStandard deviation of scores based on test preparation course:\n", prep_course_std_scores)
```

```python
#e.The management needs your help to give bonus points to the top 25% of students based on their math score,
#so how will you help the management to achieve this.

sorted_data = data.sort_values(by='math score', ascending=False)
top_25_percent = int(len(data) * 0.25)
bonus_points = 10  # Adjust bonus points as needed
top_students = sorted_data.head(top_25_percent)
top_students['bonus_points'] = bonus_points

print("Top 25% of students with bonus points:\n", top_students)
```

# **Case Study on Testing of Hypothesis**



**2.A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it.**

```python
data2 = pd.read_csv('/content/Sales_add.csv')
data2.head()
```

```python
#a.The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

from scipy.stats import ttest_rel
t_statistic, p_value = ttest_rel(data2['Sales_before_digital_add(in $)'], data2['Sales_After_digital_add(in $)'])
alpha = 0.05

if p_value < alpha:
    print("There is a significant increase in sales after implementing digital marketing (reject the null hypothesis).")
else:
    print("There is no significant increase in sales after implementing digital marketing (fail to reject the null hypothesis).")
```

```python
#b.The company needs to check whether there is any dependency between the features “Region” and “Manager”.

from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data2['Region'], data2['Manager'])
chi2, p_value, _, _ = chi2_contingency(contingency_table)
alpha = 0.05
if p_value < alpha:
    print("There is a significant dependency between 'Region' and 'Manager' (reject the null hypothesis).")
else:
    print("There is no significant dependency between 'Region' and 'Manager' (fail to reject the null hypothesis).")
```
